```python
#Ejercicio N 01
#Escribir un programa que pregunte al usuario o genere las ventas'
#de los últimos 5 días, y muestre por pantalla una serie (Dataframe)
#con los datos de las ventas por producto (opcional): Nro de
#unidades y precio unitario, indexada por los días, antes y después
#de aplicarles un descuento del 5%.

# Autores: Arraiza Nicola Erick Omar y Montes Chapilliquen Anahys Mercedes 

import pandas as pd
import numpy as np

# Preguntamos al usuario o generamos aleatoriamente las ventas
ventas = []
for i in range(5):
    dia = input(f"Ingrese las ventas del día {i+1} separadas por comas (ej: 10,20,30): ")
    ventas_dia = [int(v) for v in dia.split(",")]
    ventas.append(ventas_dia)
ventas = np.array(ventas)

# Creamos un dataframe con las ventas
df = pd.DataFrame(ventas, columns=["Producto 1", "Producto 2", "Producto 3", "Producto 4", "Producto 5"])
df.index = ["Día 1", "Día 2", "Día 3", "Día 4", "Día 5"]
print("Ventas sin descuento:\n", df)
# Aplicamos el descuento del 5%
descuento = 0.05
df_descuento = df * (1 - descuento)

# Creamos un nuevo dataframe con las ventas con descuento
df_descuento.columns = [f"{col} con descuento" for col in df_descuento.columns]
df_total = pd.concat([df, df_descuento], axis=1)
print("Ventas con descuento:\n", df_total)

# Ejercicio N° 02
#Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.
#El resultado final debe ser una serie con los estadísticos.

import pandas as pd

def obtener_estadisticos(diccionario_ingresos):
    # Convertir el diccionario a un objeto DataFrame de Pandas
    df = pd.DataFrame.from_dict(diccionario_ingresos, orient='index', columns=['ingreso'])
    
    # Calcular los estadísticos deseados
    ingreso_minimo = df['ingreso'].min()
    ingreso_maximo = df['ingreso'].max()
    ingreso_promedio = df['ingreso'].mean()
    
    # Crear una serie con los estadísticos
    serie_estadisticos = pd.Series([ingreso_minimo, ingreso_maximo, ingreso_promedio], index=['mínimo', 'máximo', 'promedio'])
    
    return serie_estadisticos

# Ejercicio 03
#Escribir programa que genere y muestre por pantalla un DataFrame
#con los datos de la siguiente tabla:

#Mes Ventas Gastos

#Enero 30500 22000

#Febrero 35600 23450

#Marzo 28300 18100

#Abril 33900 35700

#Mayo 42500 32450

#A continuación, generar nuevas columna:
#1. Columna 1 = Gasto/Ventas*100.
#2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,
#caso contrario imprimir “Meta no Superada”.
#3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
#4. Obtener el total de las ventas de los últimos 3 meses
#5. Obtener el total de gastos de febrero a abril

# Crear el diccionario con los ingresos mensuales de los clientes de una empresa
diccionario_ingresos = {
    'cliente1': 5000,
    'cliente2': 7500,
    'cliente3': 10000,
    'cliente4': 3000,
    'cliente5': 9000
}

# Obtener los estadísticos de los ingresos mensuales
serie_estadisticos = obtener_estadisticos(diccionario_ingresos)

# Imprimir los estadísticos obtenidos
print(serie_estadisticos)

import pandas as pd

# Crear el DataFrame con los datos de la tabla
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [30500, 35600, 28300, 33900, 42500],
    'Gastos': [22000, 23450, 18100, 35700, 32450]
}

df = pd.DataFrame(datos)

# Agregar la columna 1: Gasto/Ventas*100
df['Porcentaje Gasto/Ventas'] = df['Gastos']/df['Ventas']*100

# Agregar la columna 2: Meta Superada o Meta no Superada
df['Meta'] = ['Meta Superada' if venta >= 30000 else 'Meta no Superada' for venta in df['Ventas']]

# Agregar la columna 3: Sobre gasto si Gasto > Ventas
df['Sobre Gasto'] = [gasto - venta if gasto > venta else None for gasto, venta in zip(df['Gastos'], df['Ventas'])]

# Calcular el total de ventas de los últimos 3 meses (marzo, abril y mayo)
total_ventas_ultimos_3_meses = df.loc[2:, 'Ventas'].sum()

# Calcular el total de gastos de febrero a abril
total_gastos_febrero_abril = df.loc[1:3, 'Gastos'].sum()

# Imprimir el DataFrame con las nuevas columnas y los totales calculados
print(df)
print('\nTotal de ventas de los últimos 3 meses:', total_ventas_ultimos_3_meses)
print('Total de gastos de febrero a abril:', total_gastos_febrero_abril)

#Ejercicio 04
#Con el siguiente fichero datos_seguros.csv contiene información de riesgo de clientes de un banco. Escribir un programa con los siguientes requisitos:
#1. Generar un DataFrame con los datos del fichero.
#2. Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas, los tipos de datos de las columnas, mostrar las 10 primeras filas y las 10 últimas filas, verificar si tiene datos perdidos/nulos.
#3. Extraer las 1000 primeras filas y almacenarlas en un contenedor.

import pandas as pd

# Leer el archivo y crear el DataFrame
df = pd.read_csv('datos_seguros.csv')

# Mostrar información básica del DataFrame
print('Dimensiones del DataFrame:', df.shape)
print('Número de datos que contiene el DataFrame:', df.size)
print('Nombres de las columnas:', df.columns.tolist())
print('Tipos de datos de las columnas:\n', df.dtypes)
print('Las 10 primeras filas:\n', df.head(10))
print('Las 10 últimas filas:\n', df.tail(10))
print('¿Tiene datos perdidos/nulos?:', df.isnull().values.any())

# Extraer las 1000 primeras filas y almacenarlas en un contenedor
primeras_1000_filas = df[:1000]
print('Número de filas extraídas:', len(primeras_1000_filas))
```
